### Step 3:Clean the text data by removing stop words, punctuation and performing stemming/lemmatization.

#### Handle stop words & punctuation:

```python
text_column = 'reviewText'
label_column = 'Positive'
english_stopwords = set(stopwords.words('english'))
english_punctuation = string.punctuation
```

```python
def preprocess_text(text): 
    # Remove punctuation
    remove_punc = [char for char in text if char not in english_punctuation]
    clean_text = ''.join(remove_punc) # char joining

    #Remove stopwords
    words = clean_text.split()
    text = ([word for word in words if word.lower() not in english_stopwords])
    return text
```

```python
data[text_column] = data[text_column].apply(preprocess_text) 
```

```python
data[text_column] 
```

```python
data[text_column][0]
```

#### Lemmatization:

```python
lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    lemmatized_text = ' '.join([lemmatizer.lemmatize(word) for word in text])
    return lemmatized_text

data[text_column] = data[text_column].apply(lemmatize_text)
```

```python
data[text_column] 
```

```python
data[text_column][0]
```

### Step 4:Tokenize the comments.

```python
#for tokenize dataframe I use word_tokenizer
def tokenize_text(text):
    return word_tokenize(text)

data[text_column] = data[text_column].apply(tokenize_text)
```

```python
data[text_column][0]
```

#### Lemmatization:

```python
lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    lemmatized_text = ' '.join([lemmatizer.lemmatize(word) for word in text])
    return lemmatized_text

data[text_column] = data[text_column].apply(lemmatize_text)
```

### Step 5:Convert the text data into a numerical format using TF-IDF or word embeddings.

```python
#I use TF-IDF
vectorizer = TfidfVectorizer()

x = vectorizer.fit_transform(data[text_column])
y = data[label_column]
```

#### Evaluate Logistic Regression Model-

```python
from sklearn.metrics import classification_report, accuracy_score, precision_score
from sklearn.metrics import recall_score, f1_score, confusion_matrix,RocCurveDisplay
from termcolor import colored
```

```python
train_accuracy = L_reg.score(xtrain,ytrain)
test_accuracy = L_reg.score(xtest, ytest)
accuracy_score = accuracy_score(ytest, L_reg.predict(xtest))
precision_score = precision_score(ytest, L_reg.predict(xtest))
recall_score = recall_score(ytest, L_reg.predict(xtest))
f1_score = f1_score(ytest, L_reg.predict(xtest))
cm = confusion_matrix(ytest, L_reg.predict(xtest))

print(colored('Model Evaluation:\n',color = 'blue', attrs = ['bold','dark']))
print(colored(f'train_accuracy : {round(train_accuracy,2)}',color='light_magenta'))
print(colored(f'test_accuracy : {round(test_accuracy,2)}',color='light_magenta'))
print(colored(f'accuracy_score : {round(accuracy_score,2)}',color='light_magenta'))
print(colored(f'precision_score : {round(precision_score,2)}',color='light_magenta'))
print(colored(f'recall_score : {round(recall_score,2)}',color='light_magenta'))
print(colored(f'f1_score : {round(f1_score,2)}',color='light_magenta'))
print(colored(f'\nconfusion_matrix :\n {cm}',color='light_yellow'))
```
